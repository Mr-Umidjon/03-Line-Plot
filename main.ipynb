# x or y ticks

```python
x = [0, 1, 2]
y = [2, 6, 4]

plt.plot(x, y, alpha=0.7, ls='--', lw=3, marker='o', ms=10, mec='red', mfc='red')
plt.grid(ls='--', alpha=.2)
plt.xticks(ticks=x)
plt.show()
```

```python
x = [0, 1, 2]
y = [2, 6, 4]

plt.plot(x, y, alpha=0.7, ls='--', lw=3, marker='o', ms=10, mec='red', mfc='red')
plt.grid(ls='--', alpha=.2)
plt.xticks(ticks=[0, 1, 2, 3])
plt.show()
```

```python
x = [0, 1, 2]
y = [2, 6, 4]

plt.plot(x, y, alpha=0.7, ls='--', lw=3, marker='o', ms=10, mec='red', mfc='red')
plt.grid(ls='--', alpha=.2)
plt.xticks(ticks=[0, 0.5, 1.0, 1.5, 2])
plt.show()
```

```python
x = [0, 1, 2]
y = [2, 6, 4]

plt.plot(x, y, alpha=0.7, ls='--', lw=3, marker='o', ms=10, mec='red', mfc='red')
plt.grid(ls='--', alpha=.2)
plt.xticks(ticks=[0, 0.5, 1, 1.5, 2], labels=['a', 'b', 'c', 'd', 'e'])
plt.show()
```
