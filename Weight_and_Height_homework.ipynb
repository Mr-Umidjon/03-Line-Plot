### `weight-height.csv` fayilni yuklab oling.

 **Gender** - ustunini listga 0 va 1 ko'rinishda joylashtiring

 **Weight** - ustunini Pound da berilgan Kg ko'rinishda joylashtiring (1 kg = 2.205 pound).

 **Height** - ustunini inch da berilgan sm ko'rinishda listga joylashtiring (2.54 sm = 1 inch).

```python
# import library
import csv
import matplotlib.pyplot as plt
```

```python
def read_csv_data(file_path):
    """
    Read csv file
    Args:
        file_path(str): file path
    Returns:
        list: data split row
    """
    with open(file_path) as file:
        reader = csv.reader(file)
        return list(reader)
```

```python
data = read_csv_data('weight-height.csv')
```

```python
data[:5]
```

```python
def get_data(data):
    """
    Get data from list.
    Gender: Change Male to 0 and Female to 1
    Weight: Place the column in the Kg view given in Pound (1 kg = 2,205 pound).
    Height: Place the column on the list in the cm view given in inches (2.54 cm = 1 inch).
    Args:
        data(list): data split row
    Returns:
        tuple: gender, weight and height

    """
    gender = []
    weight = []
    height = []
    for row in data[1:]:
        gender.append(int(row[0] == 'Female'))
        weight.append(float(row[1]))
        height.append(float(row[2]))
    return gender, weight, height
```

```python
gender, weight, height = get_data(data)
```

```python
gender[0], weight[0], height[0]
```

# Scatter plot
**-** **Weight** va **height** ma'lumotlarni **gender** orqali ajratib alohida listlarga joylashtiring.

**-** Ma'lumotlarni tasvirlang.

```python
plt.figure(dpi=100)


def show(gender, weight, height):
    """
    Show data.
    Args:
        gender(list): 0 and 1
        weight(list): weight
        height(list): height
    Returns:
        None
    """
    w_male = []
    w_female = []

    h_male = []
    h_female = []

    for i in range(len(gender)):
        if gender[i]:
            w_female.append(weight[i])
            h_female.append(height[i])
        else:
            w_male.append(weight[i])
            h_male.append(height[i])

    plt.scatter(h_female, w_female, c='red', s=25, label='Female', alpha=0.7)
    plt.scatter(h_male, w_male, c='k', s=25, label='Male', alpha=0.7)
    plt.legend(ncol=2)
    plt.show()


show(gender, weight, height)
```

```python
show(gender, weight, height)
```
